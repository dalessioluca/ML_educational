```python
%matplotlib inline
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

sample_img = "/Users/ldalessi/Desktop/cameraman_01.jpg"

im = Image.open(sample_img)
im = im.convert('L')

img = np.array( im, dtype='float')/255
print("img.shape",img.shape)
print("image min and max",np.min(img),np.max(img))
plt.imshow(img,cmap='gray', vmin=0, vmax=1)
```

# Cropped

```python
cropped = img[50:100,140:190]
plt.imshow(cropped,cmap='gray', vmin=0, vmax=1)
```

# Check FFT and IFFT

```python
coeff = np.fft.fft2(cropped)
new_image = np.real(np.fft.ifft2(coeff))
print("new_image.shape",new_image.shape)
print("new_image min and max",np.min(new_image),np.max(new_image))
plt.imshow(new_image,cmap='gray', vmin=0, vmax=1)
```

# LOW_PASS FILTER

```python
coeff = np.fft.fft2(cropped)
abs_coeff = np.absolute(coeff)
coeff[-38:,-38:]=0
new_image = np.real(np.fft.ifft2(coeff))

print("new_image.shape",new_image.shape)
print("new_image min and max",np.min(new_image),np.max(new_image))
plt.imshow(new_image,cmap='gray', vmin=0, vmax=np.max(new_image))
```
